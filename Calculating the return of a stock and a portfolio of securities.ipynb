```python
# Create a data set that contains five columns representing five different companies
tickers = ['MSFT', 'AAPL', 'AMZN', 'GOOGL', 'GE'] 
new_data = pd.DataFrame()
new_data = yf.download(tickers, start = '2005-01-01', end='2022-12-01')['Adj Close']
```

```python
import matplotlib.pyplot as plt
```

```python
# e. Calculate simple rate of return of Microsoft for the given timeframe
MSFT['simple_return'] = (MSFT['Adj Close'] / MSFT['Adj Close'].shift(1)) - 1
print (MSFT['simple_return'])
```

```python
# d. Let's plot the data to interpret the results
MSFT['simple_return'].plot(figsize=(8,5))
plt.show()
#the most significant observations are mainly negative, in finance we are accustomed to seeing negative returns that have
#a much higher magnitude of positive returns. Usually positive returns accumulate over time and stock prices increase but
#when things do go wrong, stock prices tend to fall very quickly. An investor who is interested in buying a stock and holding
#it in the long run is mainly interested in the average rate of return that the stock will have.    
```

```python
# f. Let's calculate the average daily rate of return of the stock
avg_returns_d = MSFT['simple_return'].mean()
avg_returns_d
```

```python
# g. Let's calculate the average annual rate of return of the stock
avg_returns_a = MSFT['simple_return'].mean() * 250
avg_returns_a
```

```python
# h. Let's round the average annual rate of return of the stock to make it more presentable
print (str(round(avg_returns_a, 4) * 100) + ' %')
```

```python
# i. Calculate logarithmic return the difference is that logreturn is better when we explore the behaviour of a single stock
MSFT['log_return'] = np.log(MSFT['Adj Close'] / MSFT['Adj Close'].shift(1))
print (MSFT['log_return'])
```

```python
# j. Let's plot the data to interpret the results
MSFT['log_return'].plot(figsize=(8,5))
plt.show()
```

```python
# k. Let's calculate the average daily rate of return of the stock
avg_logreturns_d = MSFT['log_return'].mean()
avg_logreturns_d
```

```python
# l. Let's calculate the average annual rate of return of the stock
avg_logreturns_a = MSFT['log_return'].mean() * 250
avg_logreturns_a
```

```python
# m. Let's round the average annual rate of return of the stock to make it more presentable
print (str(round(avg_logreturns_a, 3) * 100) + ' %')
```

```python
# n. Calculating a portfolio of securities rate of return 
new_data.info()
```

```python
# iloc indexer with zero between the brackets will extract the data from the first column of the table
new_data.iloc[0]
```

```python
# Let's create a line chart which allow us to understand how these securities perfom throughout the time under consideration
(new_data / new_data.iloc[0] * 100).plot(figsize = (15,6));
plt.show()
#This formula help us normalize the data to 100. That's why in the graph all the lines initiate from the vertical axis
#at the point corresponding to the valua of 100. The idea of this operation is comparing the stock as if they were 
#all starting from the same value. 
```

```python
#if we don't do it, we will see the following, and we will not have a good landmark because we have different starting points
new_data.plot(figsize = (15,6))
```

```python
#We must say that Apple is a fair winner, it beats the stock price performance from 2005 to date
#If past performance is indicative of expected future performance, this type of analysis could guide your decision making
(new_data / new_data.iloc[0] * 100).plot(figsize = (15,6));
plt.show()
```

```python
# n. Calculating a portfolio of securities rate of return using simple returns. 
returns = (new_data / new_data.shift(1)) - 1
returns.head()
```

```python
#To do it let's assume equals weights for each portfolio
weights = np.array([0.25, 0.25, 0.25, 0.25, 0.25])
```

```python
np.dot(returns, weights)
```

```python
annual_returns = returns.mean() * 250
annual_returns
```

```python
np.dot(annual_returns, weights)
```

```python
pfolio_1 = (str(round(np.dot(annual_returns, weights), 3) * 100) + ' %')
print (pfolio_1)
```

```python
#To compare this combination with the performance of another portfolio, we must assign new weights 
weights_2 = np.array([0.4, 0.4, 0.10, 0.05, 0.05])
```

```python
pfolio_2 = (str(round(np.dot(annual_returns, weights_2), 3) * 100) + ' %')
print (pfolio_1)
print (pfolio_2)
#It is clear the portfolio 2 has a better return that portfolio 1
```

```python
# Create a data set that contains four columns representing four indexes
tickers = ['^GSPC', '^IXIC', '^GDAXI', '^FTSE'] 
ind_data = pd.DataFrame()
ind_data = yf.download(tickers, start = '1997-01-01', end='2022-12-01')['Adj Close']
```

```python
ind_data.info()
```

```python
#Note that the values of indices are much higher with respect to the adjusting closing price of companies
#That is only natural, given that indices are baskets composed of many stocks
ind_data.head()
```

```python
#Let's normalize the data to 100 and plot it like a graph
#Interpretation of the graph: Around of the millenium we observe a rise of the indicies value. This was because there was a dotcom bubble and the stock markets were boosted by a significance rise of tech companies.
#In the long term we can say that overall all indicies perform very well with only FTSE below the initial growth level.
#NASDAQ scores the highest in certain periods of time, experiencing in 2020 the sharpest growth.
#The German Market demonstrate real persistence and stability. 
(ind_data / ind_data.iloc[0] * 100).plot(figsize = (15,6));
plt.show()
```

```python
ind_returns = (ind_data / ind_data.shift(1)) - 1
ind_returns.tail()
```

```python
annual_ind_returns = ind_returns.mean() * 250
annual_ind_returns
#We can conclude that NASDAQ surpasses the rest of the indicies easily
#Explained by the sharp growth experimented by tech companies from the pandemic years.
```
